```python
import torch
import time
```

```python
outs = torch.rand(60,14000)
outs = torch.nn.Softmax(dim=1)(outs)
outs.shape
```

```python
sample = torch.rand(60, 1)
```

```python
outs, sample
```

```python
start = time.time()
sums = torch.zeros(outs.shape)
sums[:, 0:1] = outs[:, 0:1]
for i in range(1, outs.shape[1]):
    sums[:, i:i+1] = sums[:, i-1:i] + outs[:, i:i+1]
print(time.time() - start)
sums
```

```python
tmp = torch.arange(outs.shape[1], 0, -1)
tmp2 = torch.where(sample > sums, torch.tensor(1), torch.tensor(-1))
print(tmp * tmp2)
torch.argmin(tmp * tmp2, 1, keepdims=True)
```
