```python
import cv2
import matplotlib.pyplot as plt
import torch
import torchvision.transforms as transforms
import os
import json
import shutil
```

```python
with open("activity_net.v1-3.min.json") as f:
    data = json.load(f)
```

```python
data['database']['lTFCUuLtUJs']['annotations'][0]['label']
```

```python
def build_data(directory, root_destination, dim, collected_frames):
    annotations = []
    pid_counter = -1
    pids = []
    for action in os.listdir(directory):
        print(action)
        fullpath = os.path.join(directory, action)
        destination_action = root_destination + "/" + action
        try:
            os.mkdir(destination_action) 
        except:
            shutil.rmtree(destination_action)
            os.mkdir(destination_action)
        for pid in os.listdir(fullpath):
            if len(data['database'][pid]['annotations']) > 0:
                label = data['database'][pid]['annotations'][0]['label']
            else:
                label = ''
            pid_counter += 1
            fullpath2 = os.path.join(fullpath, pid)
            destination_action_pid = destination_action + "/" + pid
            try:
                os.mkdir(destination_action_pid) 
            except:
                shutil.rmtree(destination_action_pid) 
                os.mkdir(destination_action_pid) 
            cap = cv2.VideoCapture(fullpath2)
            num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            print(num_frames)
            annotation = "{} 1 {} {}\n".format(os.path.join(action,pid), str(collected_frames), label)
            annotations.append(annotation)
            counter = int(num_frames/collected_frames) if num_frames >= collected_frames else 1
            for i in range(0, collected_frames):
#             for i in range(num_frames):
                j = i * counter
                cap.set(1,j+1)
                ret, frame = cap.read()
                final_destination = "{}/img_{:05d}.jpg".format(destination_action_pid, i+1)
                resized_frame = cv2.resize(frame,(dim,dim))
                cv2.imwrite(final_destination, resized_frame)
    f = open("{}_annotations.txt".format(root_destination),"w") 
    f.writelines(annotations)
    f.close()
```

```python
directory = "/datasets/home/22/422/sramaswa/CSE251B/activityNetData/training"
root_destination = "train"
dim = 256
frames_to_collect = 10
build_data(directory, root_destination, dim, frames_to_collect)
```

```python
directory = "/datasets/home/22/422/sramaswa/CSE251B/activityNetData/validation"
root_destination = "val"
dim = 256
frames_to_collect = 10
build_data(directory, root_destination, dim, frames_to_collect)
```
